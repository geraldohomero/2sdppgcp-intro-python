## Pandas e suas Estruturas

O Pandas possui duas **estruturas de dados** fundamentais para representar abstrações de dados:

1. Series: array (lista) unidimensional com índices (index) e valores, para representar séries de dados. Equivalem a uma tabela de uma só coluna;
2. Dataframes: array (lista) bidimensional com linhas (representadas pelo índice) e colunas que podem conter dados de diferentes tipos. Equivale a uma tabela ou a um conjunto de Series.

## Atributos

### Series e Dataframes
Os principais atributos encontrados, tanto em Series como em Dataframes, são:

* `index`: *labels* ou etiquetas das linhas
* `shape`: uma **tupla** com o número de elementos em cada uma das dimensões (linhas, colunas);

  **Obs.:** No caso de Series, sempre será uma tupla com o segundo valor implicitamente igual a 1;
* `size`: número de elementos (linhas x colunas)
* `values`
* `dtypes`: tipo de dado armazenado (`str`, `float`, `int`);
* `ndim`: número de dimensões. Por definição, uma Serie tem `ndim` valendo 1.

### Exclusivos de Dataframes
* `columns`: um objeto complexo que guarda informações sobre as colunas, tais como seus nomes, ou *labels*.

### Acessando atributos
Atributos são acessados e funções "chamadas" (esse é o termo que usamos para invocar uma função) sobre um Dataframe da seguinte forma:

###Opções de indexação com DataFrame:

* `df.loc[rows]`: Seleciona uma única linha ou subconjunto de linhas do DataFrame por **rótulo**;
* `df.loc[:, cols]`: Seleciona uma única coluna ou subconjunto de colunas por **rótulo**;
* `df.loc[rows, cols]`: Seleciona linha(s) e coluna(s) por **rótulo**;
* `df.iloc[rows]`: Seleciona uma única linha ou subconjunto de linhas por **posição inteira**;
* `df.iloc[:, cols]` Seleciona uma única coluna ou subconjunto de colunas por **posição inteira**;
* `df.iloc[rows, cols]`: Seleciona linha(s) e coluna(s) por **posição inteira**;
* `df.at[row, col]`: Seleciona um único valor escalar por linha e coluna (números inteiros);

```python
# df.loc[rows]
df_emendas.loc[[1,2,4]]

# df.loc[rows, cols]
df_emendas.loc[[1,2,4],["Tipo de Emenda", "Nome do Autor da Emenda"]]

```

###Vamos realizar alguns agrupamentos: por ano, estado, etc

###Antes, um ajuste
`astype`: função que ajusta o tipo de dado de uma coluna

Precisamos garantir que colunas numéricas sejam interpretadas como tais pelo Pandas.

```python
# Quais são as colunas numéricas (float)?
float_cols = [col for col in df_emendas.columns if col.startswith('Valor')]

# Substituir ponto(.) por vírgula (,)
for col in float_cols:
  df_emendas[col] = df_emendas[col].str.replace(",",".")

# Vamos transformar strings em floats
df_emendas = df_emendas.astype({
    col:'float64' for col in float_cols
    })
```

###Agrupando por ano da emenda

```python
# Vamos agrupar por ano da emenda
df_agrupado = df_emendas.groupby(['Ano da Emenda'])

# Vamos aplicar soma sobre valores numéricos
df_soma = df_agrupado\
  .sum()\
  .rename(columns={'Valor Empenhado': 'Soma Valor Empenhado (Milhões R$)',
            'Valor Liquidado': 'Média Valor Liquidado (Milhões R$)'})\
  .divide(1e6)\
  .round(2)


#Mostrar os valores
df_soma
```

###Agrupandopor ano e estado

```python
df_agrupado_2 =  df_emendas.groupby(["Ano da Emenda", "UF"])
```

```python
# Vejamos a média
df_soma = df_agrupado\
  .mean()\
  .rename(columns={'Valor Empenhado': 'Soma Valor Empenhado (Milhões R$)',
            'Valor Liquidado': 'Média Valor Liquidado (Milhões R$)'})\
  .divide(1e6)\
  .round(2)
```
