## Tipos de variáveis: int, float, string, bool



*   int - Integers: Números inteiros (Z) com precisão ilimitada
*  float - Floating point numbers: Números decimais (R) precisão específica, definida pela forma como o número é representado, em termos de número de bits
* complex - Complex: Apenas a título de conhecimento. Não vem ao caso no escopo do curso
* bool - Boolean: booleanos podem ser interpretados como um subtipo de inteiros



```python
# Integers (int):

# Estamos atribuindo na variável "valor_inteiro" o valor 1
valor_inteiro_1 = 1

# Estamos atribuindo na variável "valor_inteiro" o valor 200
valor_inteiro_2 = 200

# Estamos forçando o valor 50 para se tornar um int.
valor_inteiro_3 = int("1")

# Verificação
print(f'O valor inteiro atribuído foi {valor_inteiro_1}', '\n')
print(f'O valor inteiro atribuído foi {valor_inteiro_2}', '\n')
print(f'O valor inteiro atribuído foi {valor_inteiro_3}', '\n')
```

```python
# Floating (flot)
# Estamos atribuindo na variável "flutuante_1" o valor decimal 15.20
flutuante_1 = 15.20

# Estamos atribuindo na variável "flutuante_2" o valor decimal 154.10
flutuante_2 = 154.100

# Estamos forçando o valor 200.154 para se tornar um int.
flutuante_3 = float("200.154")

print(f'O valor flutuante atribuído foi {flutuante_1}')
print(f'O valor flutuante atribuído foi {flutuante_2}')
print(f'O valor flutuante atribuído foi {flutuante_3}')
```

```python
# Complex
# Estamos atribuindo na variável "complexo_1" o valor complexo (6+5j)
complexo_1 = 6+5j

print(f'O valor complexo atribuído foi {complexo_1}')
```

```python
# Boolean (bool)

# Estamos atribuindo na variável "boolean_1" o valor Verdadeiro
boolean_1 = True

# Estamos atribuindo na variável "boolean_2" o valor Falso
boolean_2 = bool(0)

print(f'O valor booleano atribuído foi {boolean_1}')
print(f'O valor booleano atribuído foi {boolean_2}')
```

## Operadores Básicos
*   Atribuição de valor(=)
*   Comparação (>=,<=,>,<,== )
*   Operações booleanas (AND, OR, NOT)
*   Aritméticas (+, -, /, //, *, **)
*   Iteração e indexação de estruturas de dados


```python
# Operadores comparativos: Como o próprio nome diz, podemos comparar dois valores distintos, seja ele float, date, int.

# Símbolo de igualdade (==):
igualdade_1 = 5==5
igualdade_2 = 10==5

# Verificação
print(f"5 é igual a 5? {igualdade_1}") # Me retorna um valor booleano (TRUE or FALSE)
print(f"10 é igual a 5? {igualdade_2}", '\n')

# Símbolo de maior/menor:
maior_1 = 10>15
maior_2 = 15.2 >10.15
menor_1 = 40.156 < 39

# Verificação
print(f"10 é maior que 15? {maior_1}")
print(f"15.2 é maior que 10.15? {maior_2}")
print(f"40.156 é menor que 39? {menor_1}")

# Símbolo de maior/menor igual à:
maior_igual_1 = 10>=10
maior_igual_2 = 10>=15.157

# Verificação
print(f"10 é maior ou igual a 10? {maior_igual_1}")
print(f"10 é maior ou igual a 15.157? {maior_igual_2}")
```

```python
# Operadores aritméticos

# Com o símbolo de soma(+), podemos somar dois ou mais elementos.
soma_1 = 5+98
soma_2 = 56+65

# Verificação
print(f'A soma entre os valores inteiros foi {soma_1}')
print(f'A soma entre os valores inteiros foi {soma_2}')

# Com o símbolo de substração (-), podemos substrair dois ou mais elementos.
subtracao_1 = 78-34
subtracao_2 = 12-45

# Verificação
print(f'A subtracao entre os valores inteiros foi {subtracao_1}')
print(f'A subtracao entre os valores inteiros foi {subtracao_2}')

# Com o símbolo de multiplicação (*), podemos multiplicar dois ou mais elementos.
multiplicacao_1 = 3*9
multiplicacao_2 = 23.6*67.1

# Verificação
print(f'A multiplicacao entre os valores inteiros foi {multiplicacao_1}')
print(f'A multiplicacao entre os valores foi {multiplicacao_2}', '\n', '\n')

# Com o símbolo de divisão (/), podemos dividir dois ou mais elementos.
divisao_1 = 8/2
divisao_2 = 78.6/24

# Verificação
print(f'A divisao entre os valores inteiros foi {divisao_1}')
print(f'A divisao entre os valores inteiros foi {divisao_2}', '\n', '\n')

# Com o símbolo de exponenciação (**), podemos exponenciar (Completar a descrição)
exponenciacao_1 = 4**2
exponenciacao_2 = 7**5
print(f'A exponenciacao entre os valores inteiros foi {exponenciacao_1}')
print(f'A exponenciacao entre os valores inteiros foi {exponenciacao_2}', '\n', '\n')

# Com o símbolo de resto da divisão (%), podemos descobrir o resto da divisão entre dois elementos
resto_divisao_1 = 5%2
resto_divisao_2 = 6%2
print(f'A resto_divisao entre os valores inteiros foi {resto_divisao_1}')
print(f'A resto_divisao entre os valores inteiros foi {resto_divisao_2}', '\n', '\n')

divisao_inteira_1 = 7/2
divisao_inteira_2 = 56//3
print(f'A divisao inteira o entre os valores inteiros foi {divisao_inteira_1}')
print(f'A divisao inteira o entre os valores inteiros foi {divisao_inteira_2}')
```

```python
# Operadores booleanos

# AND
bool_and_1 = (7 == 7 and 10 > 5)
bool_and_2 = (7 == 7 and 10 < 5)
print(f'O resultado lógico foi: {bool_and_1}')
print(f'O resultado lógico foi: {bool_and_2}', '\n', '\n')
```

## Condição `if` (se)



*   A condição ìf` é usado para tomada de decisão em python, identado por um bloco de código.
  * `if`: indica o início de um bloco condicional
  * `elif`: caso a condição anterior não seja verdadeira, abre uma outra condição para ser avaliada;
  * `else`: se nenhuma das condições for verdadeira, executa o que estiver identado sob esta expressão.

A lógica é:

- **se** determinada condição é verdadeira:

    execute isto aqui

- **senão, se** outra condição é verdadeira:

    execute isto aqui

- **senão**:

    execute isto aqui


```python
# IF
idade = 18

# se idade>=18:
if idade >= 18:
  print('Você já pode ter carteira de motorista', '\n', '\n')


# IF e ELSE
idade_2 = 15
if idade_2 >= 18:
  print('Você já pode ter carteira de motorista', '\n', '\n')
else:
  print('Você é menor de idade. Não pode ter carteira de motorista', '\n', '\n')


# IF, ELIF e ELSE
idade_3 = None
if idade_3 < 18:
  print('Você é menor de idade. Não pode ter carteira de motorista', '\n', '\n')
elif idade_3 > 65:
  print('Sua idade é superior a 65 anos, precisará fazer o exame de renovação da carta novamente.')
else:
  print('Você já pode ter carteira de motorista', '\n', '\n')
```

##Operadores de repetição

### **for**
Itera sobre uma lista de valores, executando os comandos sob o bloco `for`

```python
## Operador for

lista = [1, 2, 3, 4, 5]
j=0
for x in lista:
  j = j+x
  print(j)
```

```python
lista_2 = ['Banana', 'Morango', 'Laranja', 'Abacaxi', 'Hortelã']
for k in lista_2:
  print(k)
```

```python
# Reaproveitando a lista acima e usando a condição if

for k in lista_2:
  if k == "Abacaxi":
    break
  print(k)
```

```python
# Trabalhando com dicionários
dicionario = {'São Paulo': 11, 'Rio de Janeiro': 21, 'Brasília': 61, 'Amazonia': 92}
for u in dicionario:
  if u == 'São Paulo':
    print(f'O DDD de {u} é {dicionario[u]}')
```

###Iterando com a função `range`

```python
# Explicitando apenas o valor final
for y in range(10):
  print(y)
```

```python
# Explpicitando valores inicial e final
for y in range(10, 30):
  print(y)
```

```python
lista_vazia = []

# Explicitando valor inicial, final e o passo (2)
for x in range(1, 10, 2):
  lista_vazia.append(x)

print(lista_vazia)
```

```python
lista_par = []
lista_impar = []

for x in range(1, 20):
  if x % 2 == 0:
    lista_par.append(x)
  else:
    lista_impar.append(x)


print(f"Lista ímpar: {lista_impar}")
print(f"Lista par: {lista_par}")
```

```python
### Operador While
valor = 0
while valor <= 10:
  valor += 1
  print(valor)
```

###**while:**

```python
while True:
  input_2 = input("O operador while é bastante útil? (sim/não): ")
  if input_2 == 'sim':
    break
```

##BÔNUS: Manipulação de strings

*   count()
*   upper()
*   lower()
*   replace()
*   strip()
*   split()
*   lstrip()
*   rstrip()
*   capitalize()
*   startswith()
*   endswith()

```python
manipulacao_string = 'Hoje vamos aprender o básico de Python'
manipulacao_string_2 = '   Hoje vamos aprender o básico de Python   '
# A função count(), contará a quantidade de vezes que uma letra ou palavra se repete dentro da string
print(manipulacao_string.count('a'))
print(manipulacao_string.count('B'), '\n', '\n')

# A função upper(), transformará tudo em Maiusculo

print(manipulacao_string.upper(), '\n', '\n')

# A função lower(), transformará tudo em minusculo
print(manipulacao_string.lower(), '\n', '\n')

# A função replace(), irá substituir a palavra que você desejar
print(manipulacao_string.replace('Brasil', 'Mundo'), '\n', '\n')

# A função strip() irá remover os espaços do inicio e do fim da string
print(manipulacao_string_2.strip(), '\n', '\n')

# A função lstrip() irá remover os espaços no início da string
print(manipulacao_string_2.lstrip(), '\n', '\n')

# A função rstrip() irá remover os espaços no final da string
print(manipulacao_string_2.rstrip(), '\n', '\n')

# A função split() irá separar as palavras pelo comando que você determinar.
print(manipulacao_string.split(" "), '\n', '\n')

# A função capitalize() irá converter o primeiro caracter em maiusculo
print(manipulacao_string.capitalize(), '\n', '\n')

# A função endswith() irá verificar se a palavra termina com a palavra que você deseja
print(manipulacao_string.endswith('Brasil'))
print(manipulacao_string.endswith('Mundo'), '\n', '\n')

# A função startswith() irá verificar se a palavra começa com a palavra que você deseja

print(manipulacao_string.startswith('A'))
print(manipulacao_string.startswith('Base'), '\n', '\n')
```
