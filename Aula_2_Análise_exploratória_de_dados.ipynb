Aplicação do ```for```!
<br>
O que quer dizer? Pra cada elemento col da lista fornecida (que poderia ter qualquer nome, mas chamamos de col), faça o seguinte
- O seguinte é: No dataframe emendas, pegue a coluna cujo nome é col e substitua todas as vírgulas por pontos

**str.replace()**

É um método específico pra colunas de texto (strings) no pandas.
Quando você faz:
``` df["col"].str.replace(",", ".", regex=False) ```

Você está dizendo:
“Dentro de cada string dessa coluna, substitua todas as vírgulas por pontos.”
<br>
*Características importantes*: Só funciona se a coluna for dtype = object/string. <br>
``` regex=False ``` significa: trate o primeiro argumento literalmente (vírgula mesmo), ou seja, não interprete como expressão regular. Fica mais rápido e mais seguro para substituições simples.
-  o pandas não entende vírgula como sendo algo que compõe floats (tipo 15,5). Então como as coisas vêm no formato brasileiro, ficar atento pra fazer essa conversão
<br>

**astype()**

É usado para converter o tipo da coluna (ou da Series) para outro tipo.
Exemplo:
```df["col"] = df["col"].astype(float)```
Converte cada valor da coluna para float.
Serve para converter para:
- int
- float
- "string"
- "category"
- bool
- outros dtypes do numpy/pandas
<br>

````python
# com o numpy
condicoes = [
    emendas["valor_empenhado"] == 0,
    (emendas["valor_empenhado"] > 0) & (emendas["valor_empenhado"] < 1_000_000),
    (emendas["valor_empenhado"] >= 1_000_000) & (emendas["valor_empenhado"] < 50_000_000),
    (emendas["valor_empenhado"] >= 50_000_000) & (emendas["valor_empenhado"] < 500_000_000),
    (emendas["valor_empenhado"] >= 500_000_000) & (emendas["valor_empenhado"] < 1_000_000_000),
    (emendas["valor_empenhado"] >= 1_000_000_000)]

valores = [
    "Sem recurso",
    "Até 1 milhão",
    "De 1 a 50 milhões",
    "De 50 mi a 500 mi",
    "De 500 milhões a 1 bilhão",
    "A partir de 1 bi"]

emendas["faixa_valor_empenhado"] = np.select(condicoes, valores, default="Indefinido") # quando tiver algo que não cabe nas condições
````

````python
# com o pd.cut
emendas["faixa_valor_empenhado2"] = pd.cut(
    emendas["valor_empenhado"],
    bins=[-0.01, 0, 1_000_000, 50_000_000, 500_000_000, 1_000_000_000, float("inf")],
    labels=[
        "Sem recurso",
        "Até 1 milhão",
        "De 1 a 50 milhões",
        "De 50 mi a 500 mi",
        "De 500 milhões a 1 bilhão",
        "A partir de 1 bi"
    ],
    right=True,
    include_lowest=True) # o pd.cut já lida com valores nulos ou que não cabem nas faixas de forma default. O que não encaixa vira nulo
````

Cada par consecutivo define um intervalo:
- -0.01 a 0
- 0 a 1M
- 1M a 50M
- 50M a 500M
- 500M a 1B
- 1B a infinito
Usei -0.01 pra incluir valor 0 na faixa "Sem recurso". Os labels seguem na ordem exata dos bins/limites intervalos.

No pd.cut, os intervalos são fechados à direita por padrão. ``` right=True ``` significa:
- lado esquerdo aberto: ( ]
- lado direito fechado: (a, b] → inclui b, mas não inclui a.
Mas dá pra controlar isso com right=False e include_lowest=False.

````python
# com uma função
def classificar_valor(coluna):
    if pd.isna(coluna):
        return "Valor ausente"

    if coluna == 0:
        return "Sem recurso"
    elif 0 < coluna < 1_000_000:
        return "Até 1 milhão"
    elif 1_000_000 <= coluna < 50_000_000:
        return "De 1 a 50 milhões"
    elif 50_000_000 <= coluna < 500_000_000:
        return "De 50 mi a 500 mi"
    elif 500_000_000 <= coluna < 1_000_000_000:
        return "De 500 milhões a 1 bilhão"
    elif coluna >= 1_000_000_000:
        return "A partir de 1 bi"
    else:
        return "Indefinido"
````

````python
emendas["faixa_valor_empenhado3"] = emendas["valor_empenhado"].apply(classificar_valor)
````

Aqui:
- Defino o nome da função e os argumentos dela. "coluna" é um placeholder pra o que eu devo incluir como argumento. No caso, essa função aceita apenas um argumento, que vai ser a coluna de um df. Aí faço os dois pontinhos pra começar as condições desta função específica.
- Já começo com "se tiver NAs na coluna, retorna 'valor_ausente'"
- Aí continuo: se a coluna tiver valor zero, retorna 'sem recurso'. Senão, se a a coluna tiver valor maior do que zero mas maior do que 1 milhão, retorna 'até 1 milhão' etc.
<br>

**elif**

1. Python testa o primeiro if. Se for verdadeiro, ele executa aquele bloco e ignora todo o resto.
2. Se o if for falso, ele passa para os elif. **elif significa “else if” -> “senão, se”**. Cada elif é uma condição alternativa mutuamente exclusiva: só vai ser testado se todos os anteriores forem falsos.
3. O primeiro elif verdadeiro vence. Quando um elif é verdadeiro o bloco dele executa, e todo o resto é ignorado (outros elif e o else).
4. Se nenhum if ou elif for verdadeiro só aí o Python cai no else.
