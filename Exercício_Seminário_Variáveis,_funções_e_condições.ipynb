# Exercício 2


*   Crie um programa em Python que permita ao usuário inserir um valor em reais (BRL) e converta esse valor para dólares (5,18) e euros (5,45) com base nas taxas de câmbio atuais. O programa deve exibir os valores convertidos para o usuário.

```python
real = float(input("Digite o seu valor em reais para a conversão: "))

dolar = round(real / 5.18, 3)
euro = round(real / 5.45, 3)

print(f"A conversão do valor para Dólar é {dolar} e em Euro é {euro}")
```

# Exercício 3


*   Desenvolva um programa em Python que permita ao usuário inserir uma temperatura em graus Celsius (°C) e, em seguida, calcule e exiba essa temperatura equivalente em graus Fahrenheit (°F) e Kelvin (K). O programa deve garantir que as conversões sejam precisas.
  * Lembre-se de que a fórmula de conversão para Fahrenheit é (°C * 9/5) + 32, e a fórmula de conversão para Kelvin é °C + 273.15.




```python
celsius = float(input("Digite a temperatura em Celsius: "))

fahrenheit = (celsius * 9 / 5 ) + 32

kelvin = (celsius + 273.15)

print(f'A temperatura em Fahrenheit é {fahrenheit} e o valor em Kelvin é {kelvin}')
```

# Exercício 4

Crie um programa em Python que funcione como uma calculadora de médias escolares e define se o aluno está aprovado, reprovado ou de recuperação, sabendo que:
* a média final é uma média ponderada entre as notas das 3 provas feitas. A primeira prova tem peso 1, a segunda e a terceira possuem peso 2 cada.
* o estudante é aprovado se a média for maior ou igual a 6
* o estudante é reprovado direto se a média for menor que 3.
* o estudante tem uma segunda chance de aprovação com a recuperação caso sua nota estiver entre 3 e 6.




```python
# defino uma função que recebe os valores das provas e retorna o status do aluno
def status_final(p1,p2,p3):
  media = (p1+2*p2+2*p3)/5
  if media < 3:
    return 'reprovado'
  elif media >= 6:
    return 'aprovado'
  else:
    return 'em recuperação'
```

```python
p1 = float(input("Digite a nota da primeira prova:"))
p2 = float(input("Digite a nota da segunda prova:"))
p3 = float(input("Digite a nota da terceira prova:"))

status_final(p1,p2,p3)
```
